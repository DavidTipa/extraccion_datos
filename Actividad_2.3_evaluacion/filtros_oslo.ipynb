```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('listings.csv')
```

```python
#filtrar los primeros 5 registros
filtro1 = df[(df["host_name"] == 'Melissa') | (df["host_name"] == 'Peter Piotr') | (df["host_name"] == 'Nina') |  (df["host_name"] == 'Nicolas') | (df["host_name"] == 'Mandy')]
filtro1
```

```python
#registros de usuarios antes de 2019
df['host_since'] = pd.to_datetime(df['host_since'], errors='coerce')
filtro2=df[df["host_since"] < "01/01/2019"]
filtro2
```

```python
#registros de usuarios que responden en un día
filtro3 = df[(df["host_response_time"] == 'within a day') | (df["host_response_time"] == 'within a few hours') | (df["host_response_time"] == 'within an hour')]
filtro3
```

```python
#registros con capacidad de alojamiento mayor o igual a 5
filtro4=df[df["accommodates"]>=5]
filtro4
```

```python
#filtrar los registros que sean de tipo "Entire home/apt" o "Private room"
filtro5 = df[(df["room_type"] == "Entire home/apt") | (df["room_type"] == "Private room")]
filtro5
```

```python
#registros con menos de 4 habitaciones
filtro6=df[df["bedrooms"]<=4]
filtro6
```

```python
# Filtrar el DataFrame para obtener registros que tengan 1 baño o sean superhost, y que sean instantáneamente reservables

filtro7 = df[(df['instant_bookable'] == 't') & ((df['bathrooms'] == 1) | (df['host_is_superhost'] == 't'))]
filtro7
```

```python
# Filtrar el DataFrame para obtener registros que tengan una calificación de limpieza menor a 3 y una calificación general mayor a 3
filtro8=df[(df["review_scores_cleanliness"]<3) & (df["review_scores_rating"]>3)]
filtro8
```
